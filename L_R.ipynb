```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([[10],[12],[15],[18],[21]])
y = np.array([10,12,15,18,21])
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
```

```python
plt.scatter(x, y, color='purple')  # Actual data points
plt.plot(x, y_pred, color='pink')  # Regression line
plt.title("Linear Regression")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
print("Intercept (Î²0):", model.intercept_)
print("Coefficient (Î²1):", model.coef_)
```

```python
print("Intercept (Î²0):", model.intercept_)
print("Coefficient (Î²1):", model.coef_)
```

```python
X = np.random.rand(80, 1) * 100  # 100 random numbers between 0 and 100
Y = 2 * X + np.random.randn(80, 1) * 10  # Linear relation with some noise
```

```python
plt.scatter(X, Y, color='green')
plt.title("Generated Dataset of 80 Numbers")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
Y_pred = model.predict(X)
```

```python
plt.scatter(X, Y, color='orange')  # Actual data points
plt.plot(X, Y_pred, color='yellow')  # Regression line
plt.title("Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
print("Intercept (Î²0):", model.intercept_)
print("Coefficient (Î²1):", model.coef_)
```
